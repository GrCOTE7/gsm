## Comprendre la clé de tout modèle : Les 4 piliers de la régression linéaire :

- Les données (X, y) : ce sont les exemples qu’on fournit à notre machine. Surface → Prix.
- Le modèle : ici une simple droite, définie par la formule f(x) = wx + b.
- La fonction coût : elle mesure à quel point notre modèle se trompe (la fameuse moyenne des erreurs au carré).
- La descente de gradient : l’algorithme magique qui ajuste w et b pour améliorer notre modèle pas à pas. C'est cela qui permet à notre machine d'apprendre
