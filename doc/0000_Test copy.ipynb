## Chrono général


```python
import time
from datetime import timedelta

def chrono(start, end,):
    elapsed = str(timedelta(seconds=end - start)).split(".")[0]
    print(f"Temps écoulé : {elapsed}")
```

```python

start = time.perf_counter()

input("Appuie sur Entrée pour arrêter le chrono...")
end = time.perf_counter()

chrono(start, end)
```

```python
print("Début du chrono...")
start = time.perf_counter()  # ou time.time()

# --- ton code à mesurer ici ---
time.sleep(1.23)
# ------------------------------

end = time.perf_counter()

chrono(start, end)
```
